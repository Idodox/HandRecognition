## Fecthing the word to create a database

### Page Download

```python
word_page = requests.get(HEBREW_WORDS_LINK)
if word_page.status_code == 200:
    print("Word Page Download Successful")
else:
    print("Word Page Download failed")

phrase_page = requests.get(HEBREW_PHRASES_LINK)
if phrase_page.status_code == 200:
    print("Phrase Page Download Successful")
else:
    print("Phrase Page Download failed")
    
english_page = requests.get(ENGLISH_WORDS_LINK)
if english_page.status_code == 200:
    print("English Page Download Successful")
else:
    print("English Page Download failed")
```
